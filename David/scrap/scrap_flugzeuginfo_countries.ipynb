```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
options = Options()
#options.headless = True

driver = webdriver.Firefox(options=options)
```

```python
url = "https://www.flugzeuginfo.net/codes_en.php#countrycodes"
```

```python
driver.get(url)
```

```python
# wait
time.sleep(3 + np.random.random(1)[0]*3)
```

```python
dict_country = {}
```

```python
alphabet = driver.find_elements_by_xpath("//table[@style='width:80%']")
for letter in alphabet:
    lines = letter.find_elements_by_xpath(".//tr")[1:]
    for line in lines:
        colonnes = line.find_elements_by_xpath(".//td")
        if colonnes:
            code = colonnes[0].text
            country = colonnes[1].text

            dict_country[code] = country
            print(code, country)
        
```

```python
# sauvegarde
df_country = pd.DataFrame(zip(dict_country.keys(), dict_country.values()))
df_country.columns = ['code','country']
df_country.to_csv('df_country.csv', sep='ยง', encoding='utf-8')
df_country.tail()
```

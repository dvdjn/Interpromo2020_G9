```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
options = Options()
#options.headless = True

driver = webdriver.Firefox(options=options)
```

```python
url = "https://www.flugzeuginfo.net/table_airportcodes_country-location_en.php"
```

```python
driver.get(url)
```

```python
# wait
time.sleep(3 + np.random.random(1)[0]*3)
```

```python
dict_icao = {}
dict_location = {}
dict_airport = {}
dict_country = {}
```

```python
alphabet = driver.find_elements_by_xpath("//table[@style='width:95%']")
for letter in alphabet:
    lines = letter.find_elements_by_xpath(".//tr")[1:]
    for line in lines:
        colonnes = line.find_elements_by_xpath(".//td")
        if colonnes:
            iata = colonnes[0].text
            icao = colonnes[1].text
            location = colonnes[2].text
            airport = colonnes[3].text
            country = colonnes[4].text

            dict_icao[iata] = icao
            dict_location[iata] = location
            dict_airport[iata] = airport
            dict_country[iata] = country
            print(iata, icao, location, airport, country)
        
```

```python
# sauvegarde
df_airport = pd.DataFrame(zip(dict_icao.keys(), dict_icao.values(), dict_location.values(), dict_airport.values(), dict_country.values()))
df_airport.columns = ['iata','icao','location','airport', 'country']
df_airport.to_csv('df_airport.csv', sep='', encoding='utf-8')
df_airport.tail()
```

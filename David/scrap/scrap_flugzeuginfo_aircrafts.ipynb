```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
options = Options()
#options.headless = True

driver = webdriver.Firefox(options=options)
```

```python
url = "https://www.flugzeuginfo.net/table_accodes_en.php"
```

```python
driver.get(url)
```

```python
# wait
time.sleep(3 + np.random.random(1)[0]*3)
```

```python
dict_manuf = {}
dict_model = {}
dict_wake = {}
```

```python
alphabet = driver.find_elements_by_xpath("//table")
len(alphabet)
```

```python
alphabet = driver.find_elements_by_xpath("//table")
for letter in alphabet:
    lines = letter.find_elements_by_xpath(".//tr")[1:]
    for line in lines:
        colonnes = line.find_elements_by_xpath(".//td")
        if colonnes:
            icao = colonnes[0].text
            manuf = colonnes[1].text
            model = colonnes[2].text
            wake = colonnes[3].text

            dict_manuf[icao] = manuf
            dict_model[icao] = model
            dict_wake[icao] = wake
            print(icao, manuf, model, wake)
        
```

```python
# sauvegarde
df_aircraft = pd.DataFrame(zip(dict_manuf.keys(), dict_manuf.values(), dict_model.values(), dict_wake.values()))
df_aircraft.columns = ['icao','manufacturer','model','wake']
df_aircraft.to_csv('df_aircraft.csv', sep='§', encoding='utf-8')
df_aircraft.tail()
```

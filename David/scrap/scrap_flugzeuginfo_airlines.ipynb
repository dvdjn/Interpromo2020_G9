```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
options = Options()
#options.headless = True

driver = webdriver.Firefox(options=options)
```

```python
url = "https://www.flugzeuginfo.net/table_airlinecodes_airline_en.php"
```

```python
driver.get(url)
```

```python
# wait
time.sleep(3 + np.random.random(1)[0]*3)
```

```python
dict_icao = {}
dict_airline = {}
dict_callsign = {}
dict_country = {}
dict_comments = {}
```

```python
alphabet = driver.find_elements_by_xpath("//table[@style='width:98%']")
for letter in alphabet:
    lines = letter.find_elements_by_xpath(".//tr")[1:]
    for line in lines:
        colonnes = line.find_elements_by_xpath(".//td")
        if colonnes:
            iata = colonnes[0].text
            icao = colonnes[1].text
            airline = colonnes[2].text
            callsign = colonnes[3].text
            country = colonnes[4].text

            dict_icao[iata] = icao
            dict_airline[iata] = airline
            dict_callsign[iata] = callsign
            dict_country[iata] = country
            print(iata, icao, airline, callsign, country)
        
```

```python
# sauvegarde
df_airline = pd.DataFrame(zip(dict_icao.keys(), dict_icao.values(), dict_airline.values(), dict_callsign.values(), dict_country.values()))
df_airline.columns = ['iata','icao','airline','callsign', 'country']
df_airline.to_csv('df_airline.csv', sep='§', encoding='utf-8')
df_airline.tail()
```

```python
import urllib.request
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
import time
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
options = Options()
#options.headless = True

driver = webdriver.Firefox(options=options)
```

```python
domaine = 'https://en.wikipedia.org'
url = f"{domaine}/wiki/List_of_airports_by_IATA_code:_A"
url
```

```python
driver.get(url)
```

```python
# wait
time.sleep(3 + np.random.random(1)[0]*3)
```

```python
liste = driver.find_elements_by_xpath("//p/a[@title='List of airports']/..")[0]
hrefs = liste.find_elements_by_xpath(".//a[contains(@title,'List of airports by')]")
liens = [href.get_attribute("href") for href in hrefs]
liens = [url] + liens
liens
```

```python
dict_icao = {}
dict_airports = {}
dict_lieux = {}
```

```python
for url in liens:
    driver.get(url)
    time.sleep(3 + np.random.random(1)[0]*3) # wait
    
    lignes = driver.find_elements_by_xpath("//table[contains(@class,'wikitable sortable')]/tbody/tr")
    for ligne in lignes:
        colonnes = ligne.find_elements_by_xpath("./td")
        if colonnes:
            iata = colonnes[0].text
            icao = colonnes[1].text
            airport = colonnes[2].find_elements_by_xpath(".//a")[0].text
            lieu = colonnes[3].text
            
            dict_icao[iata] = icao
            dict_airports[iata] = airport
            dict_lieux[iata] = lieu
            print(iata, icao, airport, lieu)
```

```python
# sauvegarde
df_airports = pd.DataFrame(zip(dict_airports.keys(), dict_icao.values(), dict_airports.values(), dict_lieux.values()))
df_airports.columns = ['iata','icao','airport','location']
df_airports.to_csv('df_airports.csv', sep='§', encoding='utf-8')
df_airports.tail()
```
